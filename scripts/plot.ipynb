# Notebook to plot training progress of agent

It will obtain data from file stored in `bullet_racecar.cma.1.32.hist.json`, where format is `env_name.optimizer.num_rollouts.popsize.hist.json`.


```python
import numpy as np

import os
import matplotlib.pyplot as plt
```

```python
file_base = 'last_experiment'
filename = os.path.join(file_base, 'statS1.npy')
title = "feature"
```

```python
data = np.load(filename)
data = data.transpose()
print(data.shape)
```

```python
fig=plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
line_max, = plt.plot(data[:, 0]/data[0, 0], data[:, 3])
line_mean, = plt.plot(data[:, 0]/data[0, 0], data[:, 2])
# line_max, = plt.plot(data[:, 0]/160, data[:, 4])
plt.legend([line_mean, line_max], ['mean','max'])
plt.title(title)
plt.xlabel('generation')
plt.ylabel('cumulative reward')
plt.savefig(file_base+".svg")
plt.show()
```

```python
fig=plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
line_max, = plt.plot(data[:, 0]/data[0, 0], data[:, 3])
line_part1, = plt.plot(data[:, 0]/data[0, 0], data[:, 4])
line_part2, = plt.plot(data[:, 0]/data[0, 0], data[:, 5])
line_part3, = plt.plot(data[:, 0]/data[0, 0], data[:, 6])
plt.legend([line_part1,line_part2,line_part3, line_max], ['part 1','part 2','part 3','sum'])
plt.title(title)
plt.xlabel('generation')
plt.ylabel('cumulative reward')
plt.savefig(file_base+"fitness_parts.svg")
plt.show()
```
